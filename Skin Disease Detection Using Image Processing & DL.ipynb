```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import PIL.Image
%matplotlib inline
```

```python

folder_path = r"G:\New folder\train"

images = []
for filename in os.listdir(folder_path):
    
    disease_folder = os.path.join(folder_path, filename)
    path = os.listdir(disease_folder)[0]
    image_path = os.path.join(disease_folder, path)
    print(f'{image_path}')
    img = pltimg.imread(image_path)
    obj = { 'disease' : filename, 'image': img}
    images.append(obj)
    
```

```python
images
```

```python
fig, ax = plt.subplots(nrows = 12, ncols = 2, figsize = (10, 30))

for i in range(len(images)):
    
    row = i // 2
    col = i % 2
    
    disease, image = images[i]['disease'], images[i]['image']
    ax[row, col].imshow(image)
    ax[row, col].axis('off')
    
plt.show()
```
